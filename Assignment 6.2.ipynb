## Question

Charts to plot:<br>
1. Create a pie chart presenting the male/female proportion<br>
2. Create a scatterplot with the Fare paid and the Age, differ the plot color by gender<br>

## Solutions :-

```python
import numpy as np
import pandas as pd
import matplotlib
matplotlib.use('nbagg')
import matplotlib.pyplot as plt
from matplotlib import cm
```

```python
titanic = pd.read_csv('https://raw.githubusercontent.com/Geoyi/Cleaning-Titanic-Data/master/titanic_original.csv', sep=',')
titanic
```

```python
sex_no = titanic.groupby('sex').count()
sex_no
```

```python
labels = 'male', 'female'
sizes = [843, 466]

fig, ax = plt.subplots()
ax.pie(sizes, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
ax.axis('equal')
plt.show()
```

```python
age = titanic['age']
fare = titanic['fare']

fig, ax = plt.subplots()
my_plot = ax.scatter(age, fare, c=titanic['sex']=='male', cmap=cm.RdYlGn)

ax.set_xlabel('Age of passengers')
ax.set_ylabel('fare')
plt.show()
```
